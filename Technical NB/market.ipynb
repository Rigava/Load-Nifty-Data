### How to Compare the markets from different region

```python
import yfinance as yf
```

```python
data =  yf.download('^NSEI',start="2010-01-01",end = None)
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.head(5)
```

```python
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data['Close'], name='Close', line=dict(color='blue')))

# Update layout
fig.update_layout(title='Chart for Market price with Time',
                  xaxis_title='Date',
                  yaxis_title='Price')
```

```python
us_data =  yf.download('^GSPC',start="2010-01-01",end = None)
us_data =  yf.download('^IXIC',start="2010-01-01",end = None)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data['Close'], name='India Market', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=us_data.index, y=us_data['Close'], name='US Market', line=dict(color='red')))

# Update layout
fig.update_layout(title='Market Comparison',
                  xaxis_title='Date',
                  yaxis_title='Price')
```

### Compare Markets with different scale using Log Returns

```python
import numpy as np
data['ret'] = np.log(data['Close']/data['Close'].shift(1))
data['cummReturn'] = (data['ret']+1).cumprod()
us_data['ret'] = np.log(us_data['Close']/us_data['Close'].shift(1))
us_data['cummReturn'] = (us_data['ret']+1).cumprod()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=data.index, y=data['cummReturn'], name='India Market', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=us_data.index, y=us_data['cummReturn'], name='US Market', line=dict(color='red')))

# Update layout
fig.update_layout(title='Market Comparison',
                  xaxis_title='Date',
                  yaxis_title='Price')
```
